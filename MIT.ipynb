# The Massachusetts Institute of Technology (MIT)

**Franz Wortha [franz-wortha]**

The Massachusetts Institute of Technology (MIT) is a private university in Cambridge, Massachusetts. Established in 1861, MIT has since then been at the center of the development of modern technology and science. Since it was founded, MIT has been ranked among the top academic institutions in the world.

The original MIT seal was created by a committee chaired by William Barton Rogers<a name="ref_1"></a>[<sup>[1]</sup>](#note_1) and it represents two figures, symbolizing the union of "Mens et Manus" (Mind and Hand) or knowledge and practice.<a name="ref_2"></a>[<sup>[2]</sup>](#note_2) The image below displays the original seal.

![MIT logo](https://upload.wikimedia.org/wikipedia/en/4/44/MIT_Seal.svg)

Interestingly, the MIT didn't adopt a logo for its graphical identity until 2003.<a name="ref_3"></a>[<sup>[3]</sup>](#note_3)
